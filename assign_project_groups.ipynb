### Load up the group signup google form
The studnets are responsible for signing themselves up for the self-formed groups via a google form currently in shared drive COGS108/Projects. Export the finished form to a Sheet and download the responses in CSV format. Place them in this directory and change the read_csv(filename) below to point at the proper file

#### Added support for the new dropdown google form format

```python
import re

def get_PID(row):
    sids = ""
    for col in row[1:6]:
        if not pd.isnull(col):
            sid =  re.findall('A[0-9]+', col) # List of SIDs
            if sid:  # Empty / None for groups with less than 5 people
                sids += sid[0]+", "
            
    return sids[:-2]

def get_names(row):
    names = ""
    for col in row[1:6]:
        if not pd.isnull(col):
            name =  col[:-10] # non-elegant string parsing
            if name:  # Empty / None for groups with less than 5 people
                for lname, fname in [name.split(",")]:
                    name = (fname.strip(' ') + ' ' + lname)
                names += name + ", "
        
        
            
    return names[:-2] # remove the extra common and space: sloppy but works

df = pd.read_csv('COGS 118A Group Signup (Responses) - Form Responses 1-2.csv')
df = df.drop('Email Address',axis='columns')
df['sids'] = df.apply(get_PID, axis=1)
df['names'] = df.apply(get_names, axis=1)
df.columns.values[6] = 'gitids'
```

```python
df
```

```python
cols = df.columns.to_list()

df.columns = cols
cols[1] = 'names'
cols[2] = 'sids'
cols[3] = 'gitids'
groups = df.loc[:,['names','sids','gitids']]
gnames = []
for group_num in groups.index:
    gnames.append('Group{:03d}'.format(group_num))
groups['group_name']=gnames
groups['n_people'] = groups.names.apply( lambda x: len(x.split(',')) )
groups = groups.set_index('group_name')
groups
```

### Github
OK time to create a github repo for each and every group. Note to do this you need to have gh CLI installed and have authenticated it as your user (see `gh auth login` documentation)

The process will create one repo for each group from the group_template.  Next we invite users to their repo. Then we checkout the project locally, rename the files, push, and delete the local copy



```python
# YOU NEED TO CHANGE THINGS HERE EACH QUARTER!!!
# you'll have to change the name of the quarter in command_add_instructors
# and also in grpsuffix

import os
import glob


command_create = 'gh repo create --template COGS118A/group_template -y COGS118A/' # + groupname
command_invite = 'gh api -X PUT -f permission=push repos/COGS118A/' # + groupname/collaborators/username
command_add_instructors = 'gh api -X PUT -f permission=admin /orgs/COGS118A/teams/Instructors_Sp22/repos/COGS118A/' # + groupname

grpsuffix = '-Sp22' #change this name suffix each quarter
```

```python
for grp in groups.index:

    gname = grp+grpsuffix 
    
    # create new repo
    cmd = command_create+gname
    ret = os.system(cmd)
    print(cmd,ret)
    
    # add instructors to have admin access to the repo
    cmd=command_add_instructors+gname
    ret = os.system(cmd)
    print(cmd,ret)
    
    # invite everyone in the group to the repo
    for uname in groups.loc[grp,'gitids'].split(','):
        cmd=command_invite+gname+'/collaborators/'+uname.strip()
        ret = os.system(cmd)
        print(cmd,ret)
        
    # get a local copy of the repo
    cmd='gh repo clone COGS118A/'+gname
    ret = os.system(cmd)
    print(cmd,ret)

    # rename the files in the repo to the groups name pattern
    for fname in glob.glob(gname+'/*groupXXX.ipynb'):
        source = fname.split('/')[1]
        target = source.split("_")[0]+gname+'.ipynb'
        cmd = ( 'cd '+gname+'; '+
                        ' '.join(['git mv',source,target]))
        ret = os.system(cmd)
        print(cmd,ret)
    
    # commit changes and push
    cmd = 'cd '+gname+'; git add -A; git commit -m "renamed files"; git push'
    ret = os.system(cmd)
    print(cmd,ret)
    
    # remove local copy of repo
    cmd = 'rm -rf '+gname
    ret = os.system(cmd)
    print(cmd,ret)
    
    # add instructors to the repo
    cmd=command_add_instructors+gname
    ret = os.system(cmd)
    print(cmd,ret)
```
