```python
import pandas as pd
import os
import glob
```

```python
groups = pd.read_csv('assigned_groups.csv').set_index('group_name')
```

```python
grpsuffix = '-Sp22' #change this name suffix each quarter
for grp in groups.index:

    gname = grp+grpsuffix
    
    # get a local copy of the repo
    cmd='gh repo clone COGS118A/'+gname
    ret = os.system(cmd)
    print(cmd,ret)
    
    # add new file to repo
    cmd='cp ../group_template/FinalProject_groupXXX.ipynb '+gname+'/FinalProject_group'+gname.split('-')[0][-3:]+'.ipynb'
    ret = os.system(cmd)
    print(cmd,ret)
    
    # commit changes and push
    cmd = 'cd '+gname+'; git add -A; git commit -m "added checkpoint"; git push'
    ret = os.system(cmd)
    print(cmd,ret)
    
    # remove local copy of repo
    cmd = 'rm -rf '+gname
    ret = os.system(cmd)
    print(cmd,ret)
```
